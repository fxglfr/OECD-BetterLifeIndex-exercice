```python
#revoir signification de theta
sample_data.plot(kind='scatter', x='Income   Household net financial wealth US Dollar', 
                 y='Life Satisfaction   Life satisfaction Average score', figsize=(13,9))
plt.axis([0, 190000, 0, 10])
X=np.linspace(0, 190000, 1000)
plt.plot(X, 2*X/100000, "r")
plt.text(40000, 2.7, r"$\theta_0 = 0$", fontsize=14, color="r")
plt.text(40000, 1.8, r"$\theta_1 = 2 \times 10^{-5}$", fontsize=14, color="r")
plt.plot(X, 8 - 2.3*X/100000, "g")
plt.text(5000, 9.1, r"$\theta_0 = 8$", fontsize=14, color="g")
plt.text(5000, 8.2, r"$\theta_1 = -2.3 \times 10^{-5}$", fontsize=14, color="g")
plt.plot(X, 6 + 1.12*X/100000, "b")
plt.text(5000, 5, r"$\theta_0 = 6$", fontsize=14, color="b")
plt.text(5000, 4, r"$\theta_1 = 1.12 \times 10^{-5}$", fontsize=14, color="b")
#save_fig('tweaking_model_params_plot')
plt.show()
```

Modéliser les données par la régression linéaire simple
https://www.math.univ-toulouse.fr/~besse/Wikistat/pdf/st-l-inf-regsim.pdf

```python
#modèle de régression linéaire simple (indice de satisfaction dépendrait seulement du revenu)
from sklearn import linear_model
lin1 = linear_model.LinearRegression()
Xsample = np.c_[sample_data['Income   Household net financial wealth US Dollar']]
Ysample = np.c_[sample_data['Life Satisfaction   Life satisfaction Average score']]
lin1.fit(Xsample, Ysample)
t0, t1 = lin1.intercept_[0], lin1.coef_[0][0]
t0, t1
```

```python
#Dataviz du modèle
sample_data.plot(kind='scatter', x='Income   Household net financial wealth US Dollar', 
                 y='Life Satisfaction   Life satisfaction Average score', figsize=(13,9), color='b')
plt.axis([2000, 190000, 4, 10])
X=np.linspace(0, 190000, 1000)
plt.plot(X, t0 + t1*X, "b")
plt.text(5000, 9.5, r"$\theta_0 = 6.056$", fontsize=14, color="b")
plt.text(5000, 9, r"$\theta_1 = 1.131 \times 10^{-5}$", fontsize=14, color="b")
#save_fig('best_fit_model_plot')
plt.show()
```

modèle de régression linéaire simple  appliqué à prédiction pour Chypre (revenu moyenne en 2017, World Inequality Database, 
https://wid.world/country/cyprus/) 

```python


cyprus_average_income_2017 = 36584
print('Revenu moyen de Chypre en 2017 :',cyprus_average_income_2017)
cyprus_predicted_life_satisfaction = lin1.predict(cyprus_average_income_2017)[0][0]
cyprus_predicted_life_satisfaction
```

```python
#afficher l'indice de satisfaction pour Chypre en 2017
sample_data.plot(kind='scatter', x='Income   Household net financial wealth US Dollar', 
                 y='Life Satisfaction   Life satisfaction Average score', figsize=(13,9), color='b', 
                 title='Prédiction 2017 : indice de satisfation pour Chypre')
plt.axis([2000, 190000, 4, 10])
X=np.linspace(0, 190000, 1000)
plt.plot(X, t0 + t1*X, "b")
plt.text(5000, 9.5, r"$\theta_0 = 6.056$", fontsize=14, color="b")
plt.text(5000, 9, r"$\theta_1 = 1.131 \times 10^{-5}$", fontsize=14, color="b")
plt.plot([cyprus_average_income_2017, cyprus_average_income_2017], [0, cyprus_predicted_life_satisfaction], "r--")
plt.text(25000, 4.5, r"Prediction = 6.47", fontsize=14, color="b")
plt.plot(cyprus_average_income_2017, cyprus_predicted_life_satisfaction, "ro")
#save_fig('cyprus_prediction_plot')
#save_fig('best_fit_model_plot')
plt.show()
```
